```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

class data_simulate:
     def __init__(self) -> None:
          self.provider = None


     def generate_provider(self,n):
        # Generate patient demographics
        names = pd.read_csv("datagenerator_seed_names.csv")  # Load a CSV file with names
        self.provider = pd.DataFrame({
            "resourceType":"Practitioner",
            "id": np.arange(1, n+1),
            "first_name": np.random.choice(names["first_name"], size=n),
            "last_name": np.random.choice(names["last_name"], size=n),
            "qualification": None,
            "telecom": None,
            "practitionerRole": None,
            # Add other fields as needed
        })


     def generate_provider_qualification(self):
    # Generate address
        qualification_list=['Physicians','Surgeons','Specialty Physicians','Physical Therapists ']
        system_list=['http://snomed.info/sct']
        for id in self.provider.index:
            coding_data =code_data = qualification = []
            coding_data.append({
                    "System": np.random.choice(system_list),
                    "display": np.random.choice(qualification_list),
                    "code":np.random.randint(11111, 55555)
                })
            self.provider.at[id, 'qualification'] = coding_data
        

     def generate_provider_org(self):
    # Generate organization
        org = pd.read_csv("datagenerator_seed_names_org.csv")  
        for id in self.provider.index:
            managingOrganization=[]
            managingOrganization.append({
                    "reference": np.random.choice(org["OrganizationName"]),
            })
            self.provider.at[id, 'practitionerRole'] = managingOrganization

     def generate_provider_telecom(self):
    # Generate organization
        for id in self.provider.index:
            telcom_data=[]
            telcom_data.append({
                    "system" : np.random.choice(["Phone", "Mobile"]),
                    "value": np.random.randint(11111111, 99999999),
                    "use": np.random.choice(["homr", "office"])
            })
            self.provider.at[id, 'telecom'] = telcom_data


     def getdata(self):           
        return self.provider

if __name__=='__main__':
    # This is mail method
    num_provider = 1
    ds = data_simulate()
    ds.generate_provider(num_provider)
    ds.generate_provider_qualification()  # Assume each patient has up to 3 conditions#
    ds.generate_provider_telecom()
    ds.generate_provider_org() 
    #print(ds.getdata())
    fhir_df=ds.getdata()
    #Generting patient record in FHIR format
    file_path = 'providers.json'
    fhir_df.to_json(file_path,orient='records',index=False)  # Specify 'records' for a list of JSON objects 
    #fhir_df.to_xml("patient.xml",root_name="provider",row_name="patient",index=False)  # Specify 'records' for a list of JSON objects 
```
